```python
def strip_hyphen(x):
    if isinstance(x, str):
        return x.replace('-', '0')
    else: 
        return x
df['Vol.'] = df['Vol.'].map(strip_hyphen)
```

```python
def remove_letters(df):
    for i in range(len(df)):
        if df['Vol.'].iloc[i][-1] == 'K':
            df['Vol.'].iloc[i] = float(df['Vol.'].iloc[i].replace('K', ''))*1000
        elif df['Vol.'].iloc[i][-1] == 'M':
            df['Vol.'].iloc[i] = float(df['Vol.'].iloc[i].replace('M', ''))*1000000
```

```python
arima = pm.auto_arima(y_train, start_p=1, start_q=1, test='adf', max_p=6, max_q=6, 
                  m=1, d=None, seasonal=True, start_P=0, D=1, trace=False, error_action='ignore',  
                  suppress_warnings=True, stepwise=False)
y_pred_arima = arima.predict(len(y_test))
```

```python
print(mean_absolute_error(y_test, y_pred_arima))
plt.figure(figsize=(10,6))
plt.plot(test.index, y_test, label='True Values')
plt.plot(test.index, y_pred_arima, label = 'Predictions')
plt.title('ARIMA Model: Actual and Prediction for SBER')
plt.legend();
```

```python
arimax = pm.auto_arima(y_train, exogenous=X_train, trace=False, error_action="ignore")
arimax.fit(y_train, exogenous=X_train)
y_pred_arimax = arimax.predict(n_periods=len(y_test), exogenous= X_test)
```

```python
print(mean_absolute_error(y_test, y_pred_arimax))
plt.figure(figsize=(10,6))
plt.plot(test.index, y_test, label='True Values')
plt.plot(test.index, y_pred_arimax, label = 'Predictions')
plt.title('ARIMAX Model: Actual and Prediction for SBER')
plt.legend();
```
